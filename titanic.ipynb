```python
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_excel('titanic3.xls')
newdata=pd.DataFrame(data)
newdata
```

```python
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Assuming 'fare' is the column containing categorical data
categorical_features = ['fare']

# Assuming 'data' is your DataFrame and 'newdata' is another DataFrame containing the 'survived' column
X_encoded = data[categorical_features].copy()

# Creating a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Applying the transformation
X_encoded = preprocessor.fit_transform(X_encoded)

# Creating the Decision Tree Classifier
clf = DecisionTreeClassifier()

# Fitting the model
clf.fit(X_encoded, newdata['survived'])

# Plotting the Decision Tree
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=preprocessor.get_feature_names_out(categorical_features))
plt.show()


```
